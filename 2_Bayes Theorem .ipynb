#### dice 2 times : independece

```python
trials = 1100
num_favourable_samples = 0    # E and F
num_conditional_samples = 0   # F

for t in range(trials):
    dice1 = np.random.choice([1,2,3,4,5,6])
    dice2 = np.random.choice([1,2,3,4,5,6])
    
    if dice1 + dice2 == 7 :
        num_conditional_samples += 1
        if  dice1 == 3:
            num_favourable_samples+= 1
conditional_probability = num_favourable_samples / num_conditional_samples
conditional_probability
            
```

```python
expected_value = 1/6
expected_value
```

## dice and coin together : independence:

```python
trials = 1000
num_favourable_samples = 0    # E and F
num_conditional_samples = 0   # F

for t in range(trials):
    coin = np.random.choice(["H","T"])
    dice = np.random.choice([1,2,3,4,5,6])
    
    if dice == 3 :
        num_conditional_samples += 1
        if  coin == "H":
            num_favourable_samples+= 1
conditional_probability = num_favourable_samples / num_conditional_samples
conditional_probability
            # P[E|F] = P[E ∩ F]/ P[F]
```

```python
expected_value = 1/2   # P[E|F] = P[E ∩ F]/ P[F]
expected_value
```

```python
trials = 1000
num_favourable_samples = 0    # E and F
num_conditional_samples = 0   # F

for t in range(trials):
    coin = np.random.choice(["H","T"])
    dice = np.random.choice([1,2,3,4,5,6])
    
    if dice == 3 and coin == "H" :
        num_favourable_samples += 1
    if coin == "H":
        num_conditional_samples += 1
conditional_probability = num_favourable_samples / num_conditional_samples
conditional_probability
            # P[F|E] = P[E ∩ F]/ P[E]
```

```python
expected_value = 1/6   # P[F|E] = P[E ∩ F]/ P[E]
expected_value
```
